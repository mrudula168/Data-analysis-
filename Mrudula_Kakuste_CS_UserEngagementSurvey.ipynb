```python
# User Engagement Survey for Malaysian Web Platforms
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load Dataset
data = pd.read_excel('UserEngagementSurvey.xlsx')
```

```python
# reading dataset
data.head()
```

```python
# Dropping columns by index positions
columns_to_drop = [0, 1]  # Index of 'Timestamp' and 'Terms & Conditions Agree'
data = data.drop(data.columns[columns_to_drop], axis=1)

print("\nDataset after dropping columns by index positions:")
data.head()
```

```python
# Retrieving column names
data.columns
```

```python
# Renaming columns
data.rename(columns={
    'What is your age?': 'age',
    'What is your gender?': 'gender',
    'What is your occupation?': 'occupation',
    'How often do you use the internet?':'internet_usage',
    'Which device do you primarily use to access the internet?':'device',
    'What is your main reason for using the internet?':'reason',
    'What types of Malaysian web platforms do you frequently visit?':'types_of_webplatforms',
    'How often do you interact with online shopping on websites?':'for_online_shopping',
    'How often do you interact with reading articles on websites?':'for_reading_articles',
    'How often do you interact with watching videos on websites?':'for_watching_videos',
    'How often do you interact with participating in online discussions on websites?':'for_online_discussions',
    'How often do you interact with social media integration on websites?':'social_media_integration',
    "How satisfied are you with the overall user experience of Malaysian web platforms? (Rate on a scale of 1-5, where 1 is 'Very Dissatisfied' and 5 is 'Very Satisfied') ":'overall_experience',
    'How likely are you to recommend Malaysian web platforms to others? (Rate on a scale of 1-5, where 1 is \'Very Unlikely\' and 5 is \'Very Likely\')':'will_recommend',
    'How accessible do you find Malaysian web platforms for people with disabilities? (Rate on a scale of 1-5, where 1 is \'Not Accessible at All\' and 5 is \'Very Accessible\')':'accessiblity_for_disable',
    'How successful are you at finding what you need on Malaysian web platforms? (Rate on a scale of 1-5, where 1 is \'Never Successful\' and 5 is \'Always Successful\')':'rating_for_finding_needs',
    "What improvements would you suggest for Malaysian web platforms to enhance your experience? ":'improvements',
    'Are there any additional features you would like to see implemented on Malaysian web platforms?':'additional_features'
}, inplace=True)

data.head()
```

```python
# Replacing column values
data.age.value_counts()
```

```python
data['age'] = data['age'].map({'Under 18': '0-18', '18 - 24': '18-24','25 - 34':'25-34','35 - 44':'35-44',
                               '45 - 54':'45-54',
                               '55 - 64':'55-64',
                              '65 and above':'65-100'})
```

```python
# sorting age
data.age.value_counts().sort_index()
```

```python
data['for_online_shopping'].value_counts()
```

```python
data['for_online_shopping'].map(type)
```

```python
# cleaning data
data['for_online_shopping'] = data['for_online_shopping'].map({'1 (Never)': 1, 2:2, 3:3, 4:4 , '5 (Always)':5})
data['for_reading_articles'] = data['for_reading_articles'].map({'1 (Never)': 1, 2:2, 3:3, 4:4 , '5 (Always)':5})
data['for_watching_videos'] = data['for_watching_videos'].map({'1 (Never)': 1, 2:2, 3:3, 4:4 , 5:5})
data['social_media_integration'] = data['social_media_integration'].map({'1 (Never)': 1, 2:2, 3:3, 4:4 , 5:5})
```

```python
data.head()
```

```python
# null values
# dealing with missing data
# percentage of null values in each columns
(data.isnull().sum()/len(data)*100).sort_values(ascending = False)
```

```python
# null values can be seen in columns like improvements and additional features
# removing null values
data['additional_features'].fillna(value='Not Specified',inplace=True)
data['improvements'].fillna(value='Not Specified',inplace=True)
```

```python
data.isnull().sum()
```

```python
# Data analysis using Data Visualization

# Responses based on age
age = data['age'].value_counts(ascending=True)
ax = sns.barplot(x=age.index, y=age.values, palette='Blues', order=age.index)
ax.set(xlabel='AGE',ylabel='COUNT', yticks=[], title='Responses based on age', frame_on=False)
ax.tick_params(length=0)
for container in ax.containers:
    ax.bar_label(container, color='black')
plt.show()
```

```python
# Responses based on gender
colors = ['#f71064', '#086ae8']
data.groupby('gender').size().plot(kind='pie', colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Responses Based On Gender')
plt.show()
```

```python
# User's Occupation
sns.countplot(x='occupation',hue='gender',data=data)
plt.title("User's Occupation")
plt.show()
```

```python
# How often user's use the internet
data['internet_usage'].value_counts()
```

```python
internet_usage = data['internet_usage'].value_counts()
plt.barh(internet_usage.index, internet_usage.values, color='#23d857', data=data)
plt.title("How often user's use the internet")
plt.show()
```

```python
data['device'].value_counts()
```

```python
# Devices used while using internet
colors = ['#FF5733', '#10b7e3','#f92134','#77d104']
data.groupby('device').size().plot(kind='pie', colors=colors, autopct='%1.1f%%', startangle=90, explode=(0.02,0.02,0.02,0.02))
plt.title('Devices used while using internet')
plt.show()
```

```python
# User Responses
fig, axarr = plt.subplots(2, 2, figsize=(15, 8))

data['for_online_shopping'].value_counts().sort_index().plot.bar(ax=axarr[0][0], fontsize=12, color='r')
axarr[0][0].set_title("For Online Shopping", fontsize=15)


data['for_reading_articles'].value_counts().sort_index().plot.bar(ax=axarr[0][1], fontsize=12, color='m')
axarr[0][1].set_title("For Reading Articles", fontsize=15)

data['for_watching_videos'].value_counts().sort_index().plot.bar(ax=axarr[1][0], fontsize=12, color='g')
axarr[1][0].set_title("For Watching Videos", fontsize=15)

data['for_online_discussions'].value_counts().sort_index().plot.bar(ax=axarr[1][1], fontsize=12, color='b')
axarr[1][1].set_title("For Online Discussions", fontsize=15)

for ax in axarr.flatten():
    total = sum([bar.get_height() for bar in ax.patches])
    if total > 0:
        for bar in ax.patches:
            height = bar.get_height()
            percentage = f"{(height / total) * 100:.1f}%"
            ax.text(bar.get_x() + bar.get_width()/2, height, percentage,
                    ha='center', va='bottom', fontsize=9, color='black')

plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=.5)
sns.despine()
```

```python
# web platform types
data['types_of_webplatforms'].value_counts()
```

```python
types_of_webplatforms = data['types_of_webplatforms'].value_counts(ascending=True)
ax = sns.barplot(x=types_of_webplatforms.index, y=types_of_webplatforms.values, palette='Paired', order=types_of_webplatforms.index)
ax.set(xlabel='Web platforms',ylabel='COUNT', yticks=[], title='Types of web platforms used', frame_on=False)
ax.tick_params(length=0)
plt.xticks(rotation=-45)
for container in ax.containers:
    ax.bar_label(container, color='black')
plt.show()
```

```python
# social media integration
data['social_media_integration'].value_counts()
```

```python
from matplotlib import style
style.use('ggplot')
social_media_integration = data['social_media_integration'].value_counts()
plt.scatter(social_media_integration.index,social_media_integration.values,color='red')
plt.title('Scatter graph for social_media_integration')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.xticks([1, 2, 3, 4, 5])
plt.legend()
plt.show()
```

```python
# overall experience of users
overall_experience = data['overall_experience'].value_counts()
plt.barh(overall_experience.index, overall_experience.values, color='#f83b49', data=data)
plt.ylabel('Ratings')
plt.xlabel('Count')
plt.title("What is the overall experience of user")
plt.show()
```

```python
# accessibility for disable
colors = ['#FF5733', '#10b7e3','#f92134','#77d104','#f8f53b']
data.groupby('accessiblity_for_disable').size().plot(kind='pie', colors=colors, autopct='%1.1f%%', startangle=90, explode=(0.02,0.02,0.02,0.02,0.02))
plt.title('Accessibility for disable')
plt.show()
```

```python
# recommendation and ratings
fig, axarr = plt.subplots(1, 2, figsize=(15, 5))

data['will_recommend'].value_counts().sort_index().plot.bar(ax=axarr[0], fontsize=12, color='skyblue')
axarr[0].set_title("Users will recommend platforms to others", fontsize=15)


data['rating_for_finding_needs'].value_counts().sort_index().plot.bar(ax=axarr[1], fontsize=12, color='g')
axarr[1].set_title("Rating for finding needs", fontsize=15)
```

```python
# improvements required

improvements = data['improvements'].value_counts()
plt.figure(figsize=(8,5))
ax = sns.barplot(x=improvements.index, y=improvements.values, palette='Spectral')
ax.set(xlabel='Improvements',ylabel='COUNT', yticks=[], title='What are the improvements required', frame_on=False)
ax.tick_params(length=0)
plt.xticks(rotation=45)
for container in ax.containers:
    ax.bar_label(container, color='black')
plt.show()
```

```python
# additional features
additional_features = data['additional_features'].value_counts()
plt.figure(figsize=(8,5))
ax = sns.barplot(x=additional_features.index, y=additional_features.values, palette='Set2')
ax.set(xlabel='Additional features',ylabel='COUNT', yticks=[], title='What are the additional features required', frame_on=False)
ax.tick_params(length=0)
plt.xticks(rotation=-45)
for container in ax.containers:
    ax.bar_label(container, color='black')
plt.show()
```
